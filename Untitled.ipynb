```python
import matplotlib.pyplot as plt

num_records = ['1000', '100000', '10000000']

one_thousand_generate = [8.614, 878.529, 88227.442]

bash_sort = [.006, .121, 16.698]

c_sort = [.003, .091, 6.927]

python_sort = [.036, .175, 15.627]

num_records_pos = [i for i, _ in enumerate(num_records)]

plt.figure(0)


plt.bar(num_records_pos, one_thousand_generate)
plt.xlabel("Number of Records")
plt.ylabel("Execution time in seconds")
plt.title("Execution time of generating a different amount of records")

plt.xticks(num_records_pos, num_records)
plt.show()


```

```python
plt.figure(1)

plt.bar(num_records_pos, bash_sort)
plt.xlabel("Number of Records")
plt.ylabel("Execution time in seconds")
plt.title("Execution time of bash sorting command on files with varying records")

plt.xticks(num_records_pos, num_records)
plt.show()
```

```python
plt.figure(2)

plt.bar(num_records_pos, c_sort)
plt.xlabel("Number of Records")
plt.ylabel("Execution time in seconds")
plt.title("Execution time of c sorting command on files with varying records")

plt.xticks(num_records_pos, num_records)
plt.show()
```

```python
plt.figure(3)

plt.bar(num_records_pos, c_sort)
plt.xlabel("Number of Records")
plt.ylabel("Execution time in seconds")
plt.title("Execution time of python sorting command on files with varying records")

plt.xticks(num_records_pos, num_records)
plt.show()
```
